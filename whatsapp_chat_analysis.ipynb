```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with VMKG.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
# pattern = r"\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}(?:\u202F|\s)?[AP]M\s-\s[\w\d]+:\s.*"
pattern = r"\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}(?:\u202F|\s)?[AP]M\s-\s"
```

```python
messages = re.split(pattern,data)[1:]
```

```python
print(messages)
```

```python
dates = re.findall(pattern,data)
```

```python
print(dates)
```

```python
df = pd.DataFrame({'user_message':messages,'message_date':dates})
#convert message_date type
df['message_date']=pd.to_datetime(df['message_date'], format="%m/%d/%y, %I:%M\u202f%p - ")

df.rename(columns={'message_date':'date'}, inplace=True)
df.head()
```

```python
# separate users and messages

users = []
messages = []
for message in df['user_message']:
    entry = re.split(r'^([^:]+):\s',message)
    if len(entry)>2: #user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'],inplace=True)

df.head()
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hours'] = df['date'].dt.hour
df['minutes'] = df['date'].dt.minute
```

```python
df['day_name'] = df['date'].dt.day_name()
```

```python
df['day_name'].value_counts()
```

```python
period = []
for hour in df[['day_name','hours']]['hours']:
    if hour==23:
        period.append(str(hour) + "-" + str('00'))
    elif hour==0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour)+"-"+str(hour+1))
```

```python
df['period'] = period
```

```python
import seaborn as sns
plt.figure(figsize=(20,6))
sns.heatmap(df.pivot_table(index='day_name',columns='period',values='message',aggfunc='count').fillna(0))
plt.yticks(rotation='horizontal')
plt.show()
```

```python
# to convert string of list into list
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
        
```

```python
def convert3(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count += 1
        else:
            break
    return L
        
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
# only extract director 
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
# Applying stemming
# ['loving','love','loved'] becomes ['love','love','love']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
         y.append(ps.stem(i))    
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

## text vectorization

```python
# we have to tell the similarity on the basis of tags
# we have to convert the text into vectors
# we will make each movie as a vector
# and in result the closest vector will be returned
```

```python
# We will use bag of words technique
# we will concatenate all the tags
# we want 5000 most common words for this we will calculate which words have more frequency
# we will ignore the stop_words(in,for,the,a,an etc)
```

### CountVectorizer
convert a collection of text documents to a metrix token counts
it returns a scipy sparse matrix.
  Number of movies = number of vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

### Each movie is a vector
### we have to calculate the distance of each movie with rest of the movies
### we will not calculate the ucluedian distance instead of we calculate cosine disatance ( means angle between two vectors)
### ucluedian distance (point to point distance ) fails in high dimensional data

### cosine_similarity is between 0 and 1

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# distance of first movie with all the movies
cosine_similarity(vectors)[0]
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:10]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    #enumerate stores the indexes of each movie
    # key=lambda x:x[1] means we are sorting on the basis of second number
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]

    for i in movie_list:
        # fetching the movie title via index
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
